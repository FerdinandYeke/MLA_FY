```python
#This adds a column of fahrenheight and uses all the values from the celsius columns and puts it into the fahrenheiht column.
new_csv_df= pd.DataFrame(csv_df)
new_csv_df["temp_f"] = new_csv_df["temp_c"]*9/5+32
#new_csv_df["Dew Point Temp (F)"] = new_csv_df["Dew Point Temp (C)"]*9/5+32
print(new_csv_df)
```

```python
#This reorders the columns with the Temp F column, where Temp F appears before Temp C
reOrdered_csv = new_csv_df
df= pd.DataFrame(reOrdered_csv)
Modded_DF = df.loc[:,["date_time","temp_f","temp_c","dew_point","relative_humidity","wind_speed","visibility","pressure","weather"]]
print(Modded_DF)
```
